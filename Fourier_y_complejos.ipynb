a)	Defina exactamente como es xT(x), que logra ser una función par.



Para que la funcion sea par tiene que ser simetrica al eje vertical y $b_{n}$ nos dara 0

```python
t = sp.Symbol('t')
y_1= -3*t - 6
y_2= 0 * t
y_3= 3 * t - 6

in_1 = y_1.integrate((t,-3,-2))
in_2 = y_2.integrate((t,-2,2))
in_3 = y_3.integrate((t,2,3))

a_0 = (1/3)*(in_1+in_2+in_3)
display(a_0)
```

```python
t = sp.Symbol('t')
n = sp.Symbol('n')

y_1= -3*t -6* sp.cos((n*sp.pi*t)/3)
y_2= 0 * t* sp.cos((n*sp.pi*t)/3)
y_3= 3 * t -6* sp.cos((n*sp.pi*t)/3)

in_1 = (1/3)*y_1.integrate((t,-3,-2))
in_2 = (1/3)*y_2.integrate((t,-2,2))
in_3 = (1/3)*y_3.integrate((t,2,3))

a_n = (in_1+in_2+in_3)
display(a_n)
```

```python
import random as random
a0, err0 = integrate.quad(lambda t: -3*t-6,-3,-2)
a1, err1 = integrate.quad(lambda t: 3*t-6,2,3)
a0 = a0+a1
err0 = err0+err1



a0 = (1/3)*a0
y=a0*0.5

z= np.arange(-12,12,0.1)
fig = plt.figure()
N = 20
i= 1
while (i<=N):
  ai, erri = integrate.quad(lambda t: (-3*t-6) * sp.cos((i*t*sp.pi)/3),-3,-2)
  ai_2, erri_2 = integrate.quad(lambda t: (3*t-6) * sp.cos((i*t*sp.pi)/3),2,3)
  ai=ai_2+ai
  ai= (1/(3))*ai
  erri=ai
  y = y+ai*np.cos((i*z*3.14159265)/3)
  plt.plot(z,y, c=(random.random(),random.random(),random.random()))
  i=i+1

plt.grid()
plt.show()


```

Grafico de numeros Complejos

```python
#Numero complejo

z = 3 - 5j

fig, ax = plt.subplots()
ax.quiver(0,0,z.real, z.imag, angles='xy', scale_units='xy', scale=1)
ax.axis([-1,4,-6,1])
ax.set_xlabel('Parte real')
ax.set_ylabel('Parte imaginaria')
plt.grid()
plt.show()
```

Grafico polar de numeros complejos

```python
import cmath
#Cmath es un modulo matematicos para numeros complejos, por esto se le agrega la "C"
#z = 3 - 5j
magnitud, angulo = cmath.polar(z)
print("Magnitud: ", magnitud)
print("Angulo: ", angulo*180/cmath.pi)
plt.polar([0,angulo],[0,magnitud], marker='o')
```
